```python
#import chives geojson data set into folium map
chives_url = (
    "https://chichives.com/geojson/chives-data-public.geojson"
)

m = folium.Map(location=(41.881832, -87.623177), zoom_start=3, tiles="cartodb positron")
folium.GeoJson(chives_url).add_to(m)

m.save("footprint.html")
```

```python
!pip install jenkspy
import jenkspy
folium.Choropleth(
    geo_data=chives_url,
    data=df,
    key_on = 'feature.properties.geoid',
    columns=["geoid","chwAirTemp"],
    fill_color="RdYlGn_r",
    fill_opacity=0.8,
    line_opacity=0.3,
    nan_fill_color="white",
    legend_name="Air Temperature",
    use_jenks=True
).add_to(m)

m.save("footprint.html")
```
